# Tree(BST)

```python
class bst:
    def __init__(self, key, lchild = None, rchild = None):
        self.key = key 
        self.lchild = None
        self.rchild = None
        
#     """- insertion: 
#     1)Tree is empty or not -> y: first node, n:find position of new node
#     2)check l-r subtree, empty or not -> y: data will be self.r-lchild, n: recursion"""
    
    def insertion(self, data):
        if self.key == data:
            return
        
    #y: first node
        if self.key is None:
            self.key = data
            return
        
    #n:find position of new node
        if self.key > data:
            if self.lchild:
                self.lchild.insertion(data)
            else:
                #make subtree with this node(data)
                self.lchild = bst(data)
        else:
            if self.key < data:
                if self.rchild:
                    self.rchild.insertion(data)
                else:
                    self.rchild = bst(data)
                    
                    
#     """
#     search algorithm:
#     """
    def search(self, data):
        if self.key == data:
            print("node is found!")
            return
        
        if self.key > data :
           # check left s.t is None or is not None
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node is not present in tree!")
        else: 
            #right_child
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node is not present in tree!")
                
#     """
#     #traversal algorithms : preorder, postorder, inorder
#     """    
    
    def preorder(self):
        #root, left, right
        print(self.key, end=" ")
        #check right-left s.t is present or not?
        
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
            
            
    def inorder(self):
        #left, root, right -> sort
        if self.lchild:
            self.lchild.inorder()
        print(self.key, end=" ")
        if self.rchild:
            self.rchild.inorder()
            
    def postorder(self):
        #left, right, root
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key, end=" ")
        
        
#     delete:
#     1) check tree is empty or not - > y, n: find node then delete it.
#     2) 2 child nodes : 1. replace the parent node with, node with largest key in left subtree or 
#                        2. replace the parent node with, node with smallest key in right subtree. <*****
    
    def delete(self, data):
        #
        if self.key is None:
            print("tree is empty!")
            return
        
        #
        if data < self.key:
            
            if self.lchild:
                self.lchild = self.lchild.delete(data)
            else:
                print("node is not present.")
                
                
        elif data > self.key:
            
            if self.rchild:
                self.rchild = self.rchild.delete(data)
            else:
                print("node is not present.")
            
        else:
            
            #delete the node which contains 0 or 1 child nodes.
            
            if self.lchild is None :
                temp = self.rchild
                self = None
                return temp
            
            if self.rchild is None :
                temp = self.lchild
                self = None
                return temp
            
    #delete the node which contains 2 child nodes.(smallest node in right s.t)
            
            #find smallest key in the right s.t
            node = self.rchild
            while node.lchild:
                node = node.lchild
            #replace and delete
            self.key = node.key
            self.rchild = self.rchild.delete(node.key)
            
        return self
    
```

```python
tree = bst(10)

node_list = [6, 3, 1, 6, 98, 3, 7]
for i in node_list:
    tree.insertion(i)
    
tree.inorder()
print()
tree.preorder()
print()
tree.postorder()
```
